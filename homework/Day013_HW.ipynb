### ① 印出最新文章的「作者」「標題」「時間」

```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://www.ptt.cc/bbs/NBA/index.html' #籃球 nba
url2 = 'https://www.ptt.cc/bbs/NBA/index1.html' #籃球 nba_最舊

# url = 'https://www.ptt.cc/bbs/EAseries/index.html' #歐美影集
# url = 'https://www.ptt.cc/bbs/KoreaDrama/index.html' #韓劇
# url = 'https://www.ptt.cc/bbs/index.html' #熱門看板
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
columns = ['作者', '看板', '標題', '時間']
```

```python
#<div class="r-ent"> #該文章
#         <a href="/bbs/KoreaDrama/M.1577358661.A.235.html">[心得] 《天空之城》後勁很足</a>

# <div class="r-list-sep"></div> #
#     <div class="r-ent">
    #    <a href="/bbs/KoreaDrama/M.1475758152.A.32B.html">[公告] 韓劇板板規 Ver.15 (18/09/05修訂) </a>

items = soup.findAll(True, {"class": re.compile(r'(r-ent|r-list-sep)')}) #findAll （返回string中所有與pattern相匹配的全部字串，返回形式為數組）,正則表達式
# items #搜尋 html 裡面 r-ent 或 r-list-sep ==

for item in range(len(items)):
#     print(items)
#     print(items[item])
#     print("------------------------")
#     print(items[item].attrs['class'])
#     print(items[item].attrs['div'])
#     print("------------------------") 
#     print(item.find_all(class_="date").text, item.find_all('a')[-1].text) # date日期 ???
    
    if 'r-list-sep' in items[item].attrs['class']: #找出 該文章 第1個 class
        try:
            r = BeautifulSoup(requests.get('https://www.ptt.cc'+items[item-1].a['href']).text, "html5lib")
            article = []
            for info in r.find_all(class_='article-meta-value'):
                article.append(info.text)
            print('作者: {}\n標題: {}\n時間: {}'.format(article[0], article[2], article[-1]))
        except:
            continue
```

```python
def GetFistPageInfo(url):
    r = requests.get(url)
    
    soup = BeautifulSoup(r.text, "html5lib")
   
    articles = []
#     article = [] #====
    
    for d in soup.find_all(class_="title"):
#         print(soup.find_all(class_="title")) #==
#         print("-----------------") #==
        try:
            r = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text, "html5lib")
            article = []
            for info in r.find_all(class_='article-meta-value'): # article-meta-value=作者
#                 print(r.find_all(class_='article-meta-value')) #===
#                 print("-----------------") #==
#                 #[<span class="article-meta-value">Price (Price)</span>, <span class="article-meta-value">NBA</span>, 
#                   #<span class="article-meta-value">[轉錄]Lyotard 對於太陽板討論串的結論</span>, <span class="article-meta-value">Sun May 16 01:54:33 2004</span>]

#                 print(info.text) #== 
                article.append(info.text) #存 到表格
            articles.append(article)
#             print("-----------------") #==
        except:
            continue
#     df1 = pd.DataFrame(article)
#     return df1
    df = pd.DataFrame(articles, columns=columns) #存 到表格
    return df
```
