### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
import datetime
import re
import pandas as pd
```

```python
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
acc1=[]
soup = BeautifulSoup(r.text, "html5lib")
today = datetime.date.today().strftime("%Y/%m/%d")

#----#1----
columns = ['日期', '分類', '標題'] #1
for d in soup.find(class_="part_list_2").find_all('h3'): # part_list_2下面所有 h3
    print(d.find(class_="date").text, d.find_all('a')[-1].text) # date日期
# #     print(d.find_all('a')) 
# #     #回傳一個 List，[ ] 是用來取 List 元素的用法。L[0] 取得第一個、L[1] 取得第二個...，L[-1] 是取出倒數第一個
# #     # [<a href="/news/20191229/1613228.htm" target="_blank"><i class="icon_type_video"></i>MBC演藝大賞P.O站華莎、全炫茂旁邊超巨大！「真實身高」竟登熱搜</a>]
# #     #print(type(d.find_all('a')))
    acc1.append([d.find(class_="date").text, d.find('em').text, d.find_all('a')[-1].text]) #放到陣列

# #----#2-----
# df = pd.DataFrame(columns=['時間', '分類', '標題']) #2
# i=0 #2
# for d in soup.find(class_="part_list_2").find_all('h3'):
#     print(d.find(class_="date").text, d.find_all('a')[-1].text) # date日期
#     df.loc[i] = [d.find(class_="date").text, d.find_all('em')[-1].text,d.find_all('a')[-1].text]
#     i+=1 #因為用loc
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
df = pd.DataFrame(acc1, columns=columns) #1
# df #1
df.sort_values('分類')
```

### ③ 哪一個類別的文章最多

```python
grouped = df.groupby('分類') #groupby 將資料依照自己要的column分組
grouped.groups # groups 看結果,,有什麼資料與資料類型
```

```python
# #----#1-----
df2 = pd.DataFrame(df.groupby('分類').size(), columns=['文章數量'])  #size每個組別統計數量
df2['排名'] = df2.rank(method='min', ascending=False) 
#rank排名, 大到小rank(method='first'), 小到大,rank(ascending=False, method='max')
df2.sort_values(by=['排名'])  
#sort_values ,对数据进行排序，用到了sort_values，by参数可以指定根据哪一列数据进行排序，
#ascending是设置升序和降序（选择多列或者多行排序要加[ ]，把选择的行列转换为列表，排序方式也可以同样的操作）。

# # #----#2-----
# df[['分類','標題']].groupby('分類').count().sort_values(['標題'], ascending=False).head(1) #輸出單向
```
