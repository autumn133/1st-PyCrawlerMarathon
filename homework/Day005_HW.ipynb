```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」 https://www.dcard.tw/f/pet?latest=false
name1="true"
name2="熱門"        
for i in range(1,3): #1到2
    if i == 2: 
        name1="false"
        name2="最新"  
        print("------------------")
        
    r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular='+name1) #true最新 ,false熱門
    response = r.text
    # response
    data = json.loads(response)#解析
    # print("共幾筆:" , len(data))
    data
    popular=[]
    for d in data:
        popular.append([d['title'],d['createdAt'],d['commentCount'],d['likeCount']]) #一行放進陣列
    popular_df = pd.DataFrame(popular,columns=['標題','貼文時間','留言人數','按讚人數']) #DataFrame列表 ,columns=指定欄標籤名稱 
    # popular_df["留言人數"])
    
    print(str(i)+ " "+name2+name1)
    print(name2+"平均留言人數: "+str(round(popular_df['留言人數'].mean(),1)))#str變為字串,round=四舍五入,mean求取均值
    print(name2+"平均按讚人數: "+str(round(popular_df['按讚人數'].mean(),1)))
```

```python
#高手寫法
def get_comment_and_count_avg(is_popular):
    r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular='+is_popular)
    response = r.text
    data = json.loads(response)
    article= []
    for item in data:
        article.append([item['title'],item['createdAt'],item['commentCount'],item['likeCount']])
    df = pd.DataFrame(article,columns=['標題','貼文時間','留言人數','按讚人數'])
    if is_popular == 'true':
        stype = '熱門'
    else:
        stype = '非熱門'
    print(stype+'平均留言人數: '+str(round(df['留言人數'].mean(),1))) #str變為字串,round=四舍五入,mean求取均值,
    print(stype+'平均按讚人數: '+str(round(df['按讚人數'].mean(),1)))
```
