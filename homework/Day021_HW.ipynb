## 2. 取出現在時間兩小時內的新聞

```python
'''
Your Code
'''
#--2---------------------------
from bs4 import BeautifulSoup
import datetime
# 打開瀏覽器
browser = webdriver.Chrome(executable_path='./chromedriver/chromedriver') 
#程式前一層目錄'../chromedriver',./chromedriver=程式相同路錄下
browser.get("https://www.ettoday.net/news/news-list.htm")

# 每個兩秒鐘自動往下滑
import time
for i in range(4):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")  #scrollTo(0, document.body.scrollHeight)=使用JS移動到網頁最下方：
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
# 取得資料，丟到 BeautifulSoup 解析
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
end_time = datetime.datetime.now()-datetime.timedelta(hours=2) 
#datetime.datetime.now()=現在時間,datetime.timedelta(hours=2)=2
end_time_str = datetime.datetime.strftime(end_time,'%Y/%m/%d %H:%M') #日期+時間,strftime返回以可讀字符串表示的當地時間
print(datetime.datetime.now())
print(end_time_str)
acount=0  #計算幾筆
for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find(class_="date").text > end_time_str:
        acount=acount+1
        print(acount, " ",d.find(class_="date").text, d.find_all('a')[-1].text)

# 關閉瀏覽器
browser.quit();
#--2---------------------------
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
'''
Your Code
'''
#--2---------------------------
from selenium import webdriver
from bs4 import BeautifulSoup
import datetime
import time
from selenium.webdriver.support.ui import Select

# 打開瀏覽器
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list-2019-12-23-0.htm")

#---前3天-----
threeDaysAgo = (datetime.datetime.now()+datetime.timedelta(days=-3)).strftime("%Y%m%d")#減3天
print("前3天: ",threeDaysAgo)
year=str(int(threeDaysAgo[:4]))
month=str(int(threeDaysAgo[4:6]))#去雙位數在轉文字
day = str(int(threeDaysAgo[6:8]))
# print(year,month,day,type(year))
time1500=threeDaysAgo+ "150000"
time1701=threeDaysAgo+ "170100"
# print(time1500,type(time1500))
time1500=datetime.datetime.strptime(time1500,'%Y%m%d%H%M%S').strftime('%Y/%m/%d %H:%M') #文字轉日期時間
time1701=datetime.datetime.strptime(time1701,'%Y%m%d%H%M%S').strftime('%Y/%m/%d %H:%M') #文字轉日期時間
# print(time1700 ,"轉日期: ",type(time1700))

#--選擇---
selectselY = Select(browser.find_element_by_id("selY")) #年
# # <select name="selY" id="selY"> 
selectselY.select_by_value(year)

selectselM = Select(browser.find_element_by_id("selM")) #月
# <select name="selM" id="selM">
selectselM.select_by_value(month)

selectselD = Select(browser.find_element_by_id("selD")) #日
# <select name="selD" id="selD">
selectselD.select_by_value(day)

browser.find_element_by_id('button').click() #查詢
# <input type="button" id="button" value="查詢" onclick="javascript:gotoformCalendar();">
#--選擇---

# 每個兩秒鐘自動往下滑

for i in range(32): #40
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")  #scrollTo(0, document.body.scrollHeight)=使用JS移動到網頁最下方：
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight),2000;")
    #print('轉:', i)
cls1="off" #結束迴圈
for i in range(30): #預防沒下拉到該時間
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")  #scrollTo(0, document.body.scrollHeight)=使用JS移動到網頁最下方：
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight),2000;")  
    
    # 取得資料，丟到 BeautifulSoup 解析
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")  
    
    for d in soup.find(class_="part_list_2").find_all('h3'):
        if d.find(class_="date").text < time1500:
            cls1="on" #結束迴圈
            break
    print('補轉:', i)
    if cls1=="on":
        break
        
for d in soup.find(class_="part_list_2").find_all('h3'):   
    if d.find(class_="date").text < time1500:
        break       
    if d.find(class_="date").text <= time1701:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
        
# # start_time_str = '2019/12/23 15:00'
# # end_time_str = '2019/12/23 17:00'
# end_time1=datetime.datetime.now()-datetime.timedelta(days=3) #3天
# end_time2=end_time1.strftime('%Y-%m-%d')+' 15:00:00'
# end_time21=end_time1.strftime('%Y-%m-%d')+' 17:00:00'

# for d in soup.find(class_="part_list_2").find_all('h3'):
#     if d.find(class_="date").text < end_time2:
#         break
#     if d.find(class_="date").text <= end_time21:
#         print(d.find(class_="date").text, d.find_all('a')[-1].text)

# 關閉瀏覽器
browser.quit();
# #--2---------------------------
```
